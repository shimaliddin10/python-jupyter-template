### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='blue', marker='o', s=100, label="Data Points")

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(5,5), fontsize=10)

# Customize the plot
plt.title("Scatter Plot with Annotations", fontsize=14)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Y Values", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```
