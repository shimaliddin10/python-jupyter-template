### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y = x ** 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create subplots (2x2 layout)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot: Line chart
axs[0, 0].plot(x, y, color='blue', label="y = x²")
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X Values")
axs[0, 0].set_ylabel("Y Values")
axs[0, 0].grid(True, linestyle='--', alpha=0.6)
axs[0, 0].legend()

# Second subplot: Bar chart
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.6)

# Third subplot: Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='blue', marker='o', s=100)
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X Values")
axs[1, 0].set_ylabel("Y Values")
axs[1, 0].grid(True, linestyle="--", alpha=0.6)

# Fourth subplot: Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, shadow=True)
axs[1, 1].set_title("Pie Chart")

# Adjust layout for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```
